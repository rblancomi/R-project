```r
# Load OncoSimulR
library(OncoSimulR)

# Pilot: as in the vignette of the package
(CRC_W <- allFitnessEffects(data.frame(parent = c("Root", "A", "B", "C"),
                                      child = c("A", "B", "C", "D"),
                                      s = 0.1,
                                      sh = -0.01,
                                      typeDep = "MN"), 
                           geneToModule = c("Root" = "Root",
                                            "A" = "APC, EPHA3, TCF7L2",
                                            "B" = "EVC2, PIK3CA, TP53", 
                                            "C" = "KRAS",
                                            "D" = "FBXW7")))
```

```r
# DAG plot
# It seem that the plot crashes if you execute it more then one time
# Try restarting the kernel (only in Jupyter notebook)
plot(CRC_W, expandModules = TRUE, autofit = TRUE)
```

```r
# Map to fitness

CRC_F <- evalAllGenotypes(CRC_W, order = FALSE, addwt = TRUE)

# Plot fitness landscape
# Not all labels are plotted, tried to change max.overlaps but did not work :c
plot(CRC_F, use_ggrepel = TRUE, max.overlaps = 200) 
```

```r
plotFitnessLandscape(CRC_F, use_ggrepel = TRUE) #The same
```
